```python
#converting '4' to '1'  as 0 = -ve , 1 = +ve
df['target'] = df['target'].replace(4, 1)
```

Stemming - reducing the word to its sroot word


```python
port_stem = PorterStemmer()
def stemming(content):
  stemmed_content = re.sub('[^a-zA-Z]', ' ', content) #removing everything which is not alphabet eg @, #,%
  stemmed_content = stemmed_content.lower()
  stemmed_content = stemmed_content.split()
  stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
  stemmed_content = ' '.join(stemmed_content)
  return stemmed_content
```

```python
df['stemmed_content'] = df['text'].apply(stemming)
```

```python
print(df['target'])
```

separating data and label


```python
X = df['stemmed_content'].values
Y = df['target'].values
```

```python
print(X)
```

```python
print(Y)
```

```python
#train test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2) #stratify is in order to distribute both data equally
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
X_train
```

```python
X_test
```

```python
#converting textual data to numeric data
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

# ML model Training

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train, Y_train)
```

```python
#accuracy score
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy Score on training data: ', training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on testing data: ' , test_data_accuracy) #77.716%
```

saving trained model


```python
import pickle
filename = 'trained_model.sav'
pickle.dump(model, open(filename, 'wb')) # wb = write binary permission only
```
